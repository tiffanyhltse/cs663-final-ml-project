```python
fund_data = pd.DataFrame(columns=['filename', 'duration', 'mean_pitch', 'mean_magnitude'])
for i in range(len(audio_list)):
    fname = '../AudioWAV/' + audio_list[i]
    y, sr = librosa.load(fname)
    pitches, magnitudes = librosa.piptrack(y=y, sr=sr)
    duration = librosa.get_duration(S=librosa.stft(y), sr=sr)
    mean_pitch = np.mean(pitches)
    mean_magnitude = np.mean(magnitudes)
    fund_data.loc[i] = [audio_list[i], duration, mean_pitch, mean_magnitude]
```

```python
emotion = []
for file in fund_data['filename']:
    if file[9] == 'A':
        emotion.append('Angry')
    elif file[9] == 'N':
        emotion.append('Neutral')
    elif file[9] == 'D':
        emotion.append('Disgust')
    elif file[9] == 'S':
        emotion.append('Sad')
    elif file[9] == 'F':
        emotion.append('Fear')
    elif file[9] == 'H':
        emotion.append('Happy')
```

```python
intensity = []
for file in fund_data['filename']:
    if file[-6:-4] == 'LO':
        intensity.append('Low')
    elif file[-6:-4] == 'HI':
        intensity.append('High')
    elif file[-6:-4] == 'MD':
        intensity.append('Medium')
    else:
        intensity.append('Unspecified')
```

```python
# It's eleven o'clock (IEO).
# That is exactly what happened (TIE).
# I'm on my way to the meeting (IOM).
# I wonder what this is about (IWW).
# The airplane is almost full (TAI).
# Maybe tomorrow it will be cold (MTI).
# I would like a new alarm clock (IWL)
# I think I have a doctor's appointment (ITH).
# Don't forget a jacket (DFA).
# I think I've seen this before (ITS).
# The surface is slick (TSI).
# We'll stop in a couple of minutes (WSI)

sentence = []
for file in fund_data['filename']:
    if file[5:8] == 'IEO':
        sentence.append('It\'s eleven o\'clock')
    elif file[5:8] == 'TIE':
        sentence.append('That is exactly what happened')
    elif file[5:8] == 'IOM':
        sentence.append('I\'m on my way to the meeting')
    elif file[5:8] == 'IWW':
        sentence.append('I wonder what this is about')
    elif file[5:8] == 'TAI':
        sentence.append('The airplane is almost full')
    elif file[5:8] == 'MTI':
        sentence.append('Maybe tomorrow it will be cold')
    elif file[5:8] == 'IWL':
        sentence.append('I would like a new alarm clock ')
    elif file[5:8] == 'ITH':
        sentence.append('I think I have a doctor\'s appointment')
    elif file[5:8] == 'DFA':
        sentence.append('Don\'t forget a jacket')
    elif file[5:8] == 'ITS':
        sentence.append('I think I\'ve seen this before')
    elif file[5:8] == 'TSI':
        sentence.append('The surface is slick')
    else:
        sentence.append('We\'ll stop in a couple of minutes')
```

```python
df2 = fund_data.assign(emotion=emotion)
df2['intensity'] = intensity
df2['sentence'] = sentence
df2
```

```python
data = df2.groupby(emotion).mean_magnitude.mean()
data
```
